将 extractor 生成的 json 转化为 excel

```python
import json
import pandas as pd

def add_row(df, feature_id, commit_ids, entity):
    r = {
        "original_mention": entity,
        "feature_id": feature_id,
        "commit_ids": commit_ids
    }
    if r not in df.to_dict(orient='records'):
        df.loc[len(df.index)] = r
    return df

with open('../data/features/features_output_20250102.json', 'r') as f:
    data = json.load(f)
    output_df = pd.DataFrame(columns=['original_mention', 'feature_id', 'commit_ids'])
    for group in data:
        feature_id = group['feature_id']
        commit_ids = group['commit_ids']
        for entity in group['entities']:
            output_df = add_row(output_df, feature_id, commit_ids, entity)
        for triple in group['triples']:
            output_df = add_row(output_df, feature_id, commit_ids, triple[0])
            if len(triple) == 3:
                output_df = add_row(output_df, feature_id, commit_ids, triple[2])
    output_df.to_excel('../data/features/features_output_20250102.xlsx', index=False)
```
